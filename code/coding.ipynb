```r
x = sort(runif(150,-12,12))
y = sin(x)+sin(x/2)+rnorm(150,0,.25)
library('locpol')
preds = locPolSmootherC(x,y,x,.5,0,gaussK)$beta0
plot(x,y)
lines(x,preds,col='red',lwd=2)
write.csv(preds,'preds1.csv')
```
